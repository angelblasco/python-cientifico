```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from IPython.html.widgets import interact
```

```python
x = np.linspace(0, 3, num=50)
y = x * np.sin(x)
plt.plot(x, y, color="red")
```

```python
#Vamos a cambiar la frecuencia de la funcion y vamos a crear un widget interactivo para cambiar
#la frecuencia de la funcion de forma dinamica.
```

```python
def pinta_funcion(freq):
    x = np.linspace(0, 3, num=200)
    y = x * np.sin(freq * x)
    plt.plot(x, y, color="red")
    plt.ylim(-5, 5) #fijo valores para que no cambie la escala.
    plt.title("Grafica de la funcion")
```

```python
pinta_funcion(2 * np.pi * 1.0)
```

```python
#Ahora creamos el widget.
```

```python
interact(pinta_funcion, freq=(2 * np.pi, 2 * np.pi * 10.0))
```
